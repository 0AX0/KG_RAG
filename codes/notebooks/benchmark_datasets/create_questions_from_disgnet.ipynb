```python

import pandas as pd

```

```python
DISEASE_GENE_DATA_PATH = "../../../data/benchmark_datasets/disgenet/disease_gene_association_with_score_1_from_disgenet.csv"
DISEASE_VARIANT_DATA_PATH = "../../../data/benchmark_datasets/disgenet/disease_variant_association_with_score_greater_than_point_9_from_disgenet.csv"

```

```python
gene_disease_df = pd.read_csv(DISEASE_GENE_DATA_PATH)
variant_disease_df = pd.read_csv(DISEASE_VARIANT_DATA_PATH)
```

```python
gene_disease_df.loc[:, "text"] = gene_disease_df["diseaseName"] + " associates " + "Gene "+ gene_disease_df["geneName"]
gene_disease_df
```

```python
variant_disease_df.loc[:, "text"] = "Variant "+ variant_disease_df["variantId"] + " associates " + variant_disease_df["diseaseName"]
variant_disease_df
```

```python
pd.concat([gene_disease_df.text, variant_disease_df.text], ignore_index=True)
```
