```python
import requests
```

```python
BASE_URI = "https://spoke.rbvi.ucsf.edu"
```

```python
def get_api_resp(END_POINT, params=None):
    URI = BASE_URI + END_POINT
    if params:
        return requests.get(URI, params=params)
    else:
        return requests.get(URI)
```

```python
END_POINT = "/api/v1/types"

result = get_api_resp(END_POINT)
data_spoke_types = result.json()

print("There are a total of {} node types in SPOKE".format(len(data_spoke_types["nodes"])))
node_types = list(data_spoke_types["nodes"].keys())

print("There are a total of {} edge types in SPOKE".format(len(data_spoke_types["edges"])))
edge_types = list(data_spoke_types["edges"].keys())

node_types_to_remove = ["DatabaseTimestamp", "Version"]
filtered_node_types = [node_type for node_type in node_types if node_type not in node_types_to_remove]
```

```python
node_type = "Disease"
attribute = "identifier"
value = "DOID:14330"

END_POINT = "/api/v1/neighborhood/{}/{}/{}".format(node_type, attribute, value)

params = {
    'node_filters' : filtered_node_types,
    'edge_filters': edge_types,
    'cutoff_Compound_max_phase': 3,
    'cutoff_Protein_source': ['SwissProt'],
    'cutoff_DaG_diseases_sources': ['knowledge', 'experiments'],
    'cutoff_DaG_textmining': 3,
    'cutoff_CtD_phase': 3,
    'cutoff_PiP_confidence': 0.7,
    'cutoff_ACTeG_level': ['Low', 'Medium', 'High']
}

result = get_api_resp(END_POINT, params=params)
node_context = result.json()
```

```python

def get_node_context(node_type, attribute, value):
    END_POINT = "/api/v1/neighborhood/{}/{}/{}".format(node_type, attribute, value)
    params = {
        'node_filters' : filtered_node_types,
        'edge_filters': edge_types,
        'cutoff_Compound_max_phase': 3,
        'cutoff_Protein_source': ['SwissProt'],
        'cutoff_DaG_diseases_sources': ['knowledge', 'experiments'],
        'cutoff_DaG_textmining': 3,
        'cutoff_CtD_phase': 3,
        'cutoff_PiP_confidence': 0.7,
        'cutoff_ACTeG_level': ['Low', 'Medium', 'High']
    }
    result = get_api_resp(END_POINT, params=params)
    node_context = result.json()
    node_name = node_context[0]["data"]["properties"]["name"]
    context_dict = {}
    context_dict["context of {}".format(node_name)] = node_context
    return context_dict
```

```python
node_type = "Disease"
attribute = "identifier"
value = "DOID:14330"

node_context = get_node_context(node_type, attribute, value)
```
